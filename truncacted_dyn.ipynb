```python
import matplotlib as mpl
mpl.rcParams['text.usetex'] = True
mpl.rcParams.update({'font.size': 20})
```

```python
pars = {
"rho" : 0.15,
"tau_syn" : 0.01,
"amp_p" : 0.08,
"amp_d" : -0.053,
"tau_p" : 0.025,
"tau_d" : 0.05,
}
```

```python
Ns = np.linspace(4,17,201)
Ns2 = np.linspace(4,33,201)

ht6 = [delta_W_th(N,0.06,**pars) for N in Ns]
htt6 = [delta_W_th_truncated(N,0.15,0.06) for N in Ns]
htnz6 = [delta_W_th_nozero(N,0.06,**pars) for N in Ns]

ht26 = [delta_W_th(N,0.026,**pars) for N in Ns2]
htt26 = [delta_W_th_truncated(N,0.15,0.026) for N in Ns2]
htnz3 = [delta_W_th_nozero(N,0.026,**pars) for N in Ns2]


fig, ax = plt.subplots(1,1)
ax.plot(Ns,ht6,label="$\hat{w}=0.06$",color="tab:blue",)
ax.plot(Ns,htt6,linestyle = "dashed",color="tab:blue")
ax.plot(Ns,htnz6,linestyle ="dotted",color="tab:blue")

ax.plot(Ns2,ht26,label="$\hat{w} = 0.026$",color="tab:orange",)
ax.plot(Ns2,htt26,linestyle = "dashed",color="tab:orange")
ax.plot(Ns2,htnz3,linestyle ="dotted",color="tab:orange")


ax.set_ylim([-0.0002,0.0011])
ax.set_xlim([4,33])
ax.legend()
plt.xlabel("$N$")
plt.ylabel("$\overline{\Delta W}(\mathrm{s}^{-1})$")
fig.savefig("truncated.png")
```
